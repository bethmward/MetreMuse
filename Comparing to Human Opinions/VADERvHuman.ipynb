```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from openpyxl import load_workbook
```

```python
# This function outputs the Net Sentiment Rate (NSR) of the single line. 
def oneline_vader(line) :
    
    # VADER scores the line
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(line)
    
    pos = score['pos']
    neg = score['neg']
    
    # Calculate NSR
    nsr = pos - neg
    
    return nsr
```

```python
# Open spreadsheet with survey results
spreadsheet = load_workbook(filename="SurveyAnswers.xlsx", data_only=True)
sheet = spreadsheet['Sentiment Results']

survey = []
# put each result in a list/tuple
for i in range(2, 102):
    survey.append(sheet["E" + str(i)].value)
```

```python
# Have VADER evaluate each line and add result to a list/tuple
file1 = open("lines.txt", 'r')

vader = []

for line in file1:
    nsr = oneline_vader(line)
    vader.append(nsr)

file1.close()
```

```python
plt.scatter(vader, survey)
plt.plot([-1,1], [-1, 1], color='black')
plt.title('Net Sentiment Rate of VADER Analysis vs Human Analysis')
plt.xlabel('VADER')
plt.ylabel('Survey')
plt.grid(True)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.rcParams["figure.figsize"] = (20,5)
plt.show()
```
