```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
```

```python
file1 = open('red_red_wine.txt', 'r')

lines = 0

for line in file1:
    b = line.strip()
    if b != "" :
        lines += 1
        
# Closing file(s)
file1.close()
```

```python
file1 = open('red_red_wine.txt', 'r')

TBnsr = np.zeros(lines + 1, dtype=float) # a 1x15 array (0:14 lines)
TBpolarity = 0

sid = SentimentIntensityAnalyzer()
Vnsr = np.zeros(lines + 1, dtype=float) # another 1x15 array (0:14 lines)
Vvec = [0,0]

lines = 0

for line in file1:
    b = line.strip()
    
    tb = TextBlob(b)
    v = sid.polarity_scores(b)
    
    if b != '' :
        lines += 1
                    
        # net sentiment rate (TextBlob)
        TBpolarity += tb.sentiment.polarity
        TBnsr[lines] = TBpolarity / lines
        
        # net sentiment rate (VADER)
        Vvec = [Vvec[0] + v['neg'], Vvec[1] + v['pos']]
        Vnsr[lines] = (Vvec[1] - Vvec[0]) / lines
    
# Closing file(s)
file1.close()
```

```python
plt.plot(TBnsr, label = 'Net Sentiment Rate (TextBlob)')
plt.plot(Vnsr, label = 'Net Sentiment Rate (VADER)')
plt.title('VADER and TextBlob NSR - "Red, Red Wine"')
plt.xlabel('Line')
plt.ylabel('Value')
plt.xticks(np.arange(0, 90, step=10))
plt.legend()
plt.grid(True)
plt.show()
```

```python
TBmean = np.sum(TBnsr)/len(TBnsr)
Vmean = np.sum(Vnsr)/len(Vnsr)

TBnsr = (TBnsr - TBmean) / np.sqrt(np.sum((TBnsr-TBmean)**2))
Vnsr = (Vnsr - Vmean) / np.sqrt(np.sum((Vnsr-Vmean)**2))
```

```python
plt.plot(TBnsr, label = 'Net Sentiment Rate (TextBlob)')
plt.plot(Vnsr, label = 'Net Sentiment Rate (VADER)')
plt.title('Normalised Net Sentiment Rate (VADER and TextBlob) - "Red, Red Wine"')
plt.xlabel('Line')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
